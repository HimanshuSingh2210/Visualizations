```python
import matplotlib.pyplot as plt
```

```python
classes=['python','r','ai','ml','ds']
class1_stu=[30,10,20,25,10]
```

```python
plt.pie(class1_stu)
plt.show()
```

```python
plt.pie(class1_stu,labels=classes)
plt.show()
```

```python
#explode we have to asign any float value to it
classes=['python','r','ai','ml','ds']
class1_stu=[30,10,20,25,10]

explode=[0,0,0,0.1,0]

plt.pie(class1_stu,labels=classes,explode=explode)
plt.show()
```

```python
classes=['python','r','ai','ml','ds']
class1_stu=[30,10,20,25,10]

explode=[0,0,0,0.1,0]
colors=['r','g','k','b','c']

plt.pie(class1_stu,labels=classes,explode=explode,colors=colors)
plt.show()
```

```python
#autopct will give us percent
#radius we have to provide float value
#startangle will give us first visual at that angle
#textprops we have to give dictionary value
#rotatelabels bydefault it is False
classes=['python','r','ai','ml','ds']
class1_stu=[30,10,20,25,10]

explode=[0,0,0,0.1,0.3]
colors=['r','g','k','b','c']
textprops={"fontsize":15}

plt.pie(class1_stu,labels=classes,explode=explode,colors=colors,autopct='%0.1f%%',
       shadow=True,radius=1.4,startangle=90,textprops=textprops,rotatelabels=True)
plt.show()
```

```python
wedgeprops={'linewidth':4,'width':3,'edgecolor':'y'}

classes=['python','r','ai','ml','ds']
class1_stu=[30,10,20,25,10]

explode=[0,0,0,0.1,0.3]
colors=['r','g','k','b','c']
textprops={"fontsize":15}


plt.pie(class1_stu,labels=classes,colors=colors,autopct='%0.1f%%',
        wedgeprops=wedgeprops,
       shadow=True,radius=1.4,startangle=90,textprops=textprops,rotatelabels=True)
plt.show()
```
