```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

# Histogram

```python
kv_student_age=np.random.randint(5,18,(30))
aps_student_age=np.random.randint(5,18,(30))
```

```python
print(kv_student_age)
print(aps_student_age)
```

```python
plt.hist(kv_student_age)
```

```python
plt.hist(kv_student_age)
plt.title('student_age')
plt.xlabel('age')
plt.ylabel('no_of_student')
```

```python
plt.hist(kv_student_age,bins=6,rwidth=0.8)
plt.title('student_age')
plt.xlabel('age')
plt.ylabel('no_of_student')
```

```python
plt.hist(kv_student_age,bins=6,rwidth=0.8)
plt.title('student_age')
plt.ylabel('age')
plt.xlabel('no_of_student')
```

```python
plt.hist(kv_student_age,bins=6,rwidth=0.8,color="r",label='kv_student_age')

plt.hist(aps_student_age,bins=6,rwidth=0.8,color="g",label='aps_student_age')
plt.title('student_age')
plt.xlabel('age')
plt.ylabel('no_of_student')
plt.legend()
```

```python
plt.hist(kv_student_age,bins=6,rwidth=0.8,color="r",label='kv_student_age')

plt.hist(aps_student_age,bins=6,rwidth=0.8,orientation='horizontal',color="g",label='aps_student_age')
plt.title('student_age')
plt.xlabel('age')
plt.ylabel('no_of_student')
plt.legend()
```

```python
plt.hist([kv_student_age,aps_student_age],bins=6,rwidth=0.8,color=["r",'g'],label=['kv_student_age','aps_student_age'])

# plt.hist(aps_student_age,bins=6,rwidth=0.8,color="g",label='aps_student_age')
plt.title('student_age')
plt.xlabel('age')
plt.ylabel('no_of_student')
plt.legend()
```
