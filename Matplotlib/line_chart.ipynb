```python
import matplotlib.pyplot as plt
```

```python
days=[]
for i in range (0,10):
    x=i+1
    days.append(x)
print(days)
```

```python
temp=[]
for i in range (30,40):
    y=i+1
    temp.append(y)
print(temp)
```

# Line Chart

```python
x=days
y=[31, 32, 33, 32, 30, 36, 37, 38, 39, 40]
```

```python
plt.plot(x,y)
```

```python
#labeling--label
#location-legend

plt.plot(x, y, color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12)

plt.title("temp chart")
plt.xlabel("days")
plt.ylabel("temp")
plt.legend('temp line',loc=4)
plt.show()
```

```python
plt.plot(x, y, color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12,label='temp line')

plt.title("temp chart")
plt.xlabel("days")
plt.ylabel("temp")
plt.legend(loc=10) #location is available from 0-10
plt.show()
```

```python
#style--style.use()
#grid--plt.grid()

from matplotlib import style
```

```python
plt.plot(x, y, color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12,label='temp line')

style.use("ggplot")

plt.title("temp chart")
plt.xlabel("days")
plt.ylabel("temp")
plt.legend(loc=4) #location is available from 0-10
plt.show()
```

```python
plt.plot(x, y, color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12,label='temp line')

style.use("ggplot")

plt.grid(color='r', linestyle='-', linewidth=2) #wanted to change colour of grid

plt.title("temp chart")
plt.xlabel("days")
plt.ylabel("temp")
plt.legend(loc=4) #location is available from 0-10
plt.show()
```

```python
y2=[29, 30, 31, 34, 38, 40, 37, 38, 34, 41]
```

```python
plt.plot(x, y, color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12,label='temp line of y')

plt.plot(x, y2, color='green', marker='^', linestyle='-',linewidth=1, markersize=12,label='temp line of y2')

style.use("ggplot")

plt.grid(color='k', linestyle='-', linewidth=2) #wanted to change colour of grid

plt.title("temp chart")
plt.xlabel("days")
plt.ylabel("temp")
plt.legend(loc=4) #location is available from 0-10
plt.show()
```
