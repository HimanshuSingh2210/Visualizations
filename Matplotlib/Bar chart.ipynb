```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes=['python','r','ai','ml','ds']
class1_stu=[30,10,20,25,10]
class2_stu=[40,5,20,20,10]
class3_stu=[35,5,30,15,15]
```

```python
plt.bar(classes,class1_stu)
```

```python
#for plotting in horizontal
plt.barh(classes,class1_stu)
```

```python
# #parameters
# plt.bar(
#     x,
#     height,
#     width=0.8,
#     bottom=None,
#     *,
#     align='center',
#     data=None,
#     **kwargs,
# )
```

```python
plt.bar(classes,class1_stu,width=0.2,align='center',color='y',edgecolor='m',linewidth=5)
```

```python
plt.bar(classes,class1_stu,align='center',color='y',edgecolor='m',linewidth=5,alpha=0.5)
#alpha can be used between 0-1
```

```python
plt.bar(classes,class1_stu,width=0.2,align='center',color='y',edgecolor='m',linewidth=5,alpha=0.8,linestyle='dashed',label='class1_stu')
```

```python
plt.bar(classes,class1_stu,width=0.2,align='center',color='y',edgecolor='m',linewidth=5,alpha=0.8,linestyle='dashed',label='class1_stu',visible=False)
#visible will help us to hide and show our data
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_stu,width=0.2,align='center',color='y',edgecolor='m',linewidth=5,alpha=1,linestyle='dashed',label='class1_stu')
style.use('ggplot')
plt.title('bar chart of courses',fontsize=18)
plt.xlabel("clases",fontsize=18)
plt.ylabel('no of stu',fontsize=19)
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_stu,width=0.2,align='center',color='y',edgecolor='m',label="class1")

plt.bar(classes,class2_stu,width=0.2,align='center',color='k',edgecolor='m',label='class2')

plt.legend()
plt.title('bar chart of courses',fontsize=18)
plt.xlabel("clases",fontsize=18)
plt.ylabel('no of stu',fontsize=19)
```

```python
plt.figure(figsize=(16,9))

classes_index=np.arange(len(classes))

width=0.2

plt.bar(classes_index,class1_stu,align='center',color='y',edgecolor='m',label="class1")

plt.bar(classes_index,class2_stu,align='center',color='k',alpha=0.5,edgecolor='m',label='class2')

plt.legend()
plt.title('bar chart of courses',fontsize=18)
plt.xlabel("clases",fontsize=18)
plt.ylabel('no of stu',fontsize=19)
```

```python


classes_index=np.arange(len(classes))

width=0.2

plt.bar(classes_index,class1_stu,align='center',color='y',edgecolor='m',label="class1")

plt.bar(classes_index+width,class2_stu,align='center',color='k',edgecolor='m',label='class2')

plt.legend()
plt.title('bar chart of courses',fontsize=18)
plt.xlabel("clases",fontsize=18)
plt.ylabel('no of stu',fontsize=19)
```

```python
#xticks--for setting the below label right
#rotation

classes_index=np.arange(len(classes))

width=0.2

plt.bar(classes_index,class1_stu,align='center',color='y',edgecolor='m',label="class1")

plt.bar(classes_index+width,class2_stu,align='center',color='k',edgecolor='m',label='class2')

plt.xticks(classes_index+width,classes,rotation=90)

plt.legend()
plt.title('bar chart of courses',fontsize=18)
plt.xlabel("clases",fontsize=18)
plt.ylabel('no of stu',fontsize=19)

```

```python


classes_index=np.arange(len(classes))

width=0.2

plt.bar(classes_index,class1_stu,width,color='y',edgecolor='m',label="class1")

plt.bar(classes_index+width,class2_stu,width,color='k',edgecolor='m',label='class2')

plt.bar(classes_index+width+width,class3_stu,width,color='r',edgecolor='m',label='class2')

plt.xticks(classes_index+width,classes,rotation=90)

plt.legend()
plt.title('bar chart of courses',fontsize=18)
plt.xlabel("clases",fontsize=18)
plt.ylabel('no of stu',fontsize=19)

```

```python


classes_index=np.arange(len(classes))

width=0.2

plt.barh(classes_index,class1_stu,width,color='y',edgecolor='m',label="class1")

plt.barh(classes_index+width,class2_stu,width,color='k',edgecolor='m',label='class2')

plt.barh(classes_index+width+width,class3_stu,width,color='r',edgecolor='m',label='class2')

plt.yticks(classes_index+width,classes,rotation=90)

plt.legend()
plt.title('bar chart of courses',fontsize=18)
plt.ylabel("clases",fontsize=18)
plt.xlabel('no of stu',fontsize=19)

```
