# if we want to give our own annotation

```python
arr = np.arange(1,13).reshape(4,3)
arr
```

```python
annot_arr = np.array([['a00','a01','a02'],
                      ['a10','a11','a12'],
                      ['a20','a21','a22'],
                      ['a30','a31','a32']])
```

```python
sns.heatmap(arr, annot= annot_arr,fmt='s')
```

```python
annot_kws={'fontsize':10, 
           'fontstyle':'italic',  
           'color':"k",
           'alpha':1, 
           'rotation':"vertical",
           'verticalalignment':'center',
#            'backgroundcolor':'g'
          }

sns.heatmap(new,cmap='coolwarm',annot=True,annot_kws=annot_kws)
```

```python
#we want to plot in seprate boxes
sns.heatmap(new,cmap='coolwarm',annot=True,linewidths=4)
```

```python
sns.heatmap(new,cmap='coolwarm',annot=True,linewidths=4,linecolor='y')
```

```python
#if we dont want the labels
plt.figure(figsize=(16,9))

sns.heatmap(new, xticklabels = False, yticklabels = False)
```

```python
plt.figure(figsize=(16,9))

cbar_kws=cbar_kws = {"orientation":"horizontal", 
            "shrink":1,
            'extend':'min', 
            'extendfrac':0.1, 
            "ticks":np.arange(0,22), 
            "drawedges":True}
                     
sns.heatmap(new,cbar_kws=cbar_kws)
```

```python
#we customise as per our terms

country_code = ['USA', 'GBR', 'IND', 'CHN', 'RUS', 'AUS', 'FRA', 'DEU', 'CAN', 'BRA', 'ARG', 'PAK', 'NPL', 'BGD', 'JPN']
 

sns.heatmap(new, xticklabels = np.arange(0,15), yticklabels = country_code)
```

```python
sns.heatmap(new, xticklabels = np.arange(0,15), yticklabels = country_code)

plt.title('Heatmap of  who is responsible for global warming')
```
