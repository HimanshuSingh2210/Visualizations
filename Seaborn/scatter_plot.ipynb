```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
titanic_df=sns.load_dataset('titanic')
```

```python
titanic_df
```

```python
sns.scatterplot(data=titanic_df,x='age',y='fare')
```

```python
sns.scatterplot(data=titanic_df,x='age',y='fare',hue='sex',style='who')
```

```python
sns.scatterplot(data=titanic_df,x='age',y='fare',hue='sex',style='who',
              size='who',sizes=(100,400))
```
