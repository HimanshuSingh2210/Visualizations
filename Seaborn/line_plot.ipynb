```python
tips_df=sns.load_dataset('tips')
```

```python
tips_df
```

```python
tips_df.shap
```

```python
sns.lineplot(x="tip",y='total_bill',data=tips_df)
plt.show()
```

```python
sns.lineplot(x="size",y='total_bill',data=tips_df)
plt.show()
```

```python
#hue-plotting with respect to
#here we are plotting with respect to sex
sns.lineplot(x="size",y='total_bill',data=tips_df,hue='sex',style="sex")
plt.show()
```

```python
# palette-coloring
# dashes True bydefault
sns.lineplot(x="size",y='total_bill',data=tips_df,hue='sex',style="sex",
             palette='Accent',dashes=False,markers=None)
plt.show()
```

```python
# legend-bydefault brief

sns.lineplot(x="size",y='total_bill',data=tips_df,hue='sex',style="sex",
             palette='Accent',markers='o')
plt.show()
```

```python
#if we want to change the default label in seaborn
```

```python
sns.lineplot(x="size",y='total_bill',data=tips_df,hue='sex',style="sex",
             palette='Accent',markers='o')

plt.title("Line_plot")
plt.xlabel('Size')
plt.ylabel('Total_bill')
```

```python
#if we want to change the background
sns.set(style='darkgrid')
        
plt.figure(figsize=(16,9))        
sns.lineplot(x="size",y='total_bill',data=tips_df,hue='sex',style="sex",
             palette='Accent',markers='o')

plt.title("Line_plot")
plt.xlabel('Size')
plt.ylabel('Total_bill')
```

```python
sns.set(style='darkgrid')
        
     
sns.lineplot(x="size",y='total_bill',data=tips_df,hue='day',style="day",
             palette='Accent',markers='o')
plt.legend(fontsize=15)

plt.title("Line_plot",fontsize=15)
plt.xlabel('Size',fontsize=15)
plt.ylabel('Total_bill',fontsize=15)
```
