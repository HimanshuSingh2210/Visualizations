```python
# hist bydefault=True but when it is False the we get only KDE(kernel density estimator)/line plot
sns.distplot(Tips['total_bill'],bins=10,hist=False)
```

```python
# kde bydefault=True but when it is False the we get only histogram
sns.distplot(Tips['total_bill'],bins=10,kde=False)
```

```python

sns.distplot(Tips['total_bill'],bins=10,rug=True)
```

```python
from scipy.stats import norm
```

```python
# norm is used for just normalise plot unde fit
sns.distplot(Tips['total_bill'],bins=10,fit=norm)
```

```python
# remove kde and just want norm

sns.distplot(Tips['total_bill'],bins=10,fit=norm,kde=False)
```
